## What major has the lowest unemployment rate? And which has the highest?

```python
min_emp = majorsdf_raw['Unemployment_rate'].min()
majorsdf_raw.loc[majorsdf_raw.Unemployment_rate == min_emp]
```

### The majors with the lowest unemployment rate are Mathmatics and Computer Science, Military Technologies, Botany, Soil Science, and Educational Administration and Supervision

## Which majors have the highest unemployment rate?

```python
max_emp = majorsdf_raw['Unemployment_rate'].max()
majorsdf_raw.loc[majorsdf_raw.Unemployment_rate == max_emp]
```

### The major with the highest Employment rate (about 18%) is Nuclear Engineering

## What about highest and lowest median salaries of these majors?

```python
min_sal = majorsdf_raw['Median'].min()
majorsdf_raw.loc[majorsdf_raw.Median == min_sal]
```

### The Lowest Median Salary major is Library Science, with the median salary of 22,000 dollars

## And what is the majors with the highest median salary?

```python
max_sal = majorsdf_raw['Median'].max()
majorsdf_raw.loc[majorsdf_raw.Median == max_sal]
```

### The Highest Median Salary is Petroleum Engineering, with the median salary of 110,000 dollars

## Which major category overall has highest percentage of Noncollege jobs out of the jobs reported?

### First, what are the categories?

```python
majorsdf_raw.Major_category.unique()
```

### How many majors are in each category?

```python
majorsdf_raw['Major_category'].value_counts()

# geeksforgeeks helped me remember how to use value_counts
# https://www.geeksforgeeks.org/how-to-extract-the-value-names-and-counts-from-value_counts-in-pandas/
# forever grateful to geeksforgeeks
```

### Now I'm making a new dataframe to make this easier for me to understand.

### This dataframe doesn't have the individual majors but it has the major categories. It also has the sum of all the jobs related to their degree and jobs not related to their degree per major category.

```python
majorsdf2 = majorsdf_raw.groupby('Major_category')[['Non_college_jobs', 'College_jobs']].sum()
majorsdf2
```

### Now I'm going to make a column with the percentage of noncollege jobs per major categories to be in the dataframe.

```python
majorsdf2['Non_percentage'] = majorsdf2['Non_college_jobs'] / (majorsdf2['Non_college_jobs'] + majorsdf2['College_jobs'])
majorsdf2
```

```python
majorsdf2.sort_values('Non_percentage')
```

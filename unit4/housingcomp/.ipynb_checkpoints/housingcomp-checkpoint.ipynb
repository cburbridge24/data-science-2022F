## Housing Kaggle Competition

#### Importing

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

train_data_raw = pd.read_csv("train.csv", index_col='Id') 
test_data = pd.read_csv("test.csv", index_col='Id') 
```

```python
train_data_raw
```

#### Making Train and Valid X and y

```python
train_data = train_data_raw.select_dtypes(include='number')
train_data.dropna(inplace=True)
test_data.dropna(inplace=True)
train_data.dtypes
```

```python
train_data1 = train_data.drop(['SalePrice', 'LotFrontage', 'MasVnrArea', 'GarageYrBlt'], axis=1)
```

```python
X = train_data[list(train_data.columns)]
y = train_data.SalePrice
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
home_model = DecisionTreeRegressor(random_state=1)
```

```python
home_model.fit(train_X, train_y)
```

```python
val_predictions = home_model.predict(val_X)
val_mae = mean_absolute_error(val_predictions, val_y)
val_mae
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 100, 500, 700, 1000, 2000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

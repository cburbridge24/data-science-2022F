### Just getting nums out of there

```python
train_data2 = train_data_raw.select_dtypes(include='number')
test_data2 = test_data_raw.select_dtypes(include='number')
```

```python
extracted_col = train_data["Transported"]

train_data2 = train_data2.join(extracted_col)
```

```python
X = train_data2[list(train_data2.columns)]
X.drop(columns='Transported', inplace=True)
y = train_data2.Transported
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=1)
```

```python
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```
